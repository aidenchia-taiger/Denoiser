```python
import cv2
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
np.set_printoptions(100)
```

```python
IMG_DIR = 'imgs/marcus_front.jpg'
```

```python
def getCluster(img):
    img = np.array(img)
    shape = img.shape
    img = img.flatten()
    img = img.reshape(-1, 1)
    gmm = GaussianMixture(n_components=2, verbose=1)
    cluster = gmm.fit_predict(img).reshape(shape)
    return cluster
```

```python
img = cv2.imread(IMG_DIR, cv2.IMREAD_GRAYSCALE)
print(img.shape)
```

```python
def display(img):
    plt.rcParams["figure.figsize"] = [12, 9]
    plt.imshow(img,'gray')
    plt.show()
    
print(np.unique(img))
display(img)
```

```python
cluster = getCluster(img)
```

```python
print(cluster)
```

```python
def getClusterMeanPI(img, cluster):
    img = np.array(img)
    numZeros = 0
    grp0_mean = 0
    numOnes = 0
    grp1_mean = 0
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if cluster[i][j] == 0:
                grp0_mean += img[i][j]
                numZeros += 1
                
            elif cluster[i][j] == 1:
                grp1_mean += img[i][j]
                numOnes += 1
    try:
        grp0_mean /= numZeros
    except:
        pass
    try:
        grp1_mean /= numOnes
    except:
        pass
    
    return grp0_mean, grp1_mean
```

```python
grp0_mean, grp1_mean = getClusterMeanPI(img, cluster)
print(grp0_mean, grp1_mean)
```

```python
def crop(img_path, height, width):
    img = Image.open(img_path).convert('L')
    imgwidth, imgheight = img.size
    crops = {}
    idx = 0
    for i in range(0,imgheight,height):
        height = imgheight - i if i+height > imgheight else height
        for j in range(0,imgwidth,width):
            width = imgwidth - j if j+width > imgwidth else width
            box = (j, i, j+width, i+height)
            crops[idx] = box
            idx += 1
    return crops
```

```python
crops = crop(IMG_DIR, 21, 21)
print(crops)
```

```python
def deshadow(img, crops, g):
    alphas = []
    for bbox in list(crops.values()):
        sub_img = img[bbox[0]:bbox[2], bbox[1]:bbox[3]]
        l = np.mean(np.array(sub_img))
        alpha = min(l / g, 1)
        img[bbox[0]:bbox[2], bbox[1]:bbox[3]] = sub_img / alpha
        
    #deshadowed = cv2.threshold(img, 125.5442379342696, 255, cv2.THRESH_BINARY)[-1]
    deshadowed = img
    return Image.fromarray(deshadowed, 'L')
```

```python
deshadowed = deshadow(img, crops, 185.24212718803454)
print(np.array(deshadowed))
```
